```python
import matplotlib.pyplot as plt
plt.plot([0,1,2,3,4], [0,1000,1300,1900,2500], 'ro-')
plt.title("Ingresos anuales")
plt.xlabel("años")
plt.ylabel("ingresos")
plt.grid(True)
plt.legend(['Ingresos'])
plt.xlim(0, 10)
plt.ylim(0, 3000)
#plt.text(5, 15, "Important Point")
```

```python
import matplotlib.pyplot as plt
plt.title("peliculasmás taquilleras 2026 ")
plt.bar(['Michael','Super Mario','Avatar','Zootopia 2'], [314,831,777,1700], color=["red","green","blue","pink","pink"])
plt.title("peliculas con mayores ingresos 2026")
plt.xlabel("peliculas")
plt.ylabel("Ganancias(Millones)")
plt.legend(['Data Series'])
plt.xlim(0, 4)
plt.ylim(0, 2000)
```

```python
import matplotlib.pyplot as plt
plt.title("Análisis de Pareto: Principales causas de rechazo en línea de producción")
plt.bar([' Barniz','Deformaciones','Fallas en sellado'], [10,70,20])
plt.ylabel("Cant de casos")
plt.xlabel("Causas")
plt.legend(['Porcentaje de las causas'])
plt.ylim(0, 109)
plt.plot([0,1,2], [10,80,100], 'ro-')
```

```python
import matplotlib.pyplot as plt
plt.title("My Awesome Plot")
plt.scatter(1, 3, c='red', marker='x')
plt.scatter(5, 8, c='blue', marker='x')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend(['Data Series'])
```

```python
import matplotlib.pyplot as plt
plt.title("My Awesome Plot")
plt.hist(0, bins=[0,0.4], color='skyblue', edgecolor='black', label="a")
plt.hist(0.5, bins=[0.4, 0.7], color='red', edgecolor='black', label="b")
plt.hist(1, bins=[0.7, 1.0], color='green', edgecolor='black', label="c")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend(['Data Series'])
plt.xlim(0, 1)
plt.ylim(0, 2)
```
